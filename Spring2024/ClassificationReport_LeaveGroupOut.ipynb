```python
#Load the Parkinson's dataset
#Read more and download at https://archive.ics.uci.edu/dataset/174/parkinsons
#also available at url = 'https://github.com/olcaykursun/ML/blob/main/Spring2024/MaxLittleParkinsons.csv'

import pandas as pd

MaxLittle_Parkinsons_Dataset = pd.read_csv('MaxLittleParkinsons.csv')

MaxLittle_Parkinsons_Dataset.head()
```

```python
MaxLittle_Parkinsons_Dataset.describe()
```

```python
#All column names
#name contains the subject number, status is the class-label
#Positive class is 1, which means Parkinson's

MaxLittle_Parkinsons_Dataset.columns
```

```python
#Remove the two aforementioned columns to obtain the predictor columns
feature_names = MaxLittle_Parkinsons_Dataset.columns.difference(['status', 'name'])

X = MaxLittle_Parkinsons_Dataset[feature_names]
y = MaxLittle_Parkinsons_Dataset['status']
groups = MaxLittle_Parkinsons_Dataset['name']

print(type(y))
print(y.shape)
print(type(X))
X
```

```python
#Subject 01 has 5 recordings of phoneme-R01
#Some subjects have 6 recordings
#Splitting 5 (or 6) recordings into train-test does not make much sense,
#because if we know recordings belong to a parkinson patient,
#then why should we try to predict the class of the other recordings of the same patient?
#Maybe for reproducibility, but ideally we would like this system for telediagnosis.

groups
```

```python
# Assuming 'groups' is a Pandas Series containing strings like 'phon_R01_S01_1'

# Extract the subject number using regex and grouping (with parentheses around the digits)
extracted_groups = groups.str.extract(r'S(\d+)_')

# Convert to NumPy array
# We can keep it as a Pandas Series, but I wanted to convert to NumPy
groups_numpy = extracted_groups.to_numpy()

# Convert to integer
# We can keep it as strings, but I wanted to convert to int
groups_int = groups_numpy.astype(int)

# Flatten the array
# We can keep it as a 2D array, but I wanted to flatten it to a 1D array.
groups_flattened = groups_int.flatten()

print(groups_flattened)
```

```python
#Obtain the data matrix
X = MaxLittle_Parkinsons_Dataset[feature_names]
print(X.head())
```

```python
#iloc is used to select rows and/or columns by integer location
X_small = X.iloc[:, 5:16]
X_small = X_small.iloc[:, [1, 3]]
X_small.head()
```

```python
#practice with indexing

import numpy as np
arr = np.arange(5,16)
some_features = arr[[1, 3]]
print(some_features)
feature_names[some_features]
```

```python
#Regular train-test split and 1-NN
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.5, stratify=y)

from sklearn.neighbors import KNeighborsClassifier

neigh = KNeighborsClassifier(n_neighbors=1)
neigh.fit(X_train, y_train)
y_pred = neigh.predict(X_test)
```

```python
#1-NN accuracy
print(neigh.score(X_test, y_test))
print(np.mean(y_pred == y_test))
n_test = y_test.shape[0]
print(f'{n_test=}')
```

```python
# The GroupShuffleSplit is used here to ensure that all recordings from a single subject
# are included entirely in either the training set or the test set, but not in both.
# This is important to prevent data leakage and ensure that the model's evaluation is fair
# and representative of its performance on unseen subjects. This is particularly crucial in studies
# where data from the same subject may be highly correlated, and learning from some
# of their data may unfairly aid in predicting the rest in testing.

from sklearn.model_selection import GroupShuffleSplit
gss = GroupShuffleSplit(n_splits=1, test_size=0.5, random_state=42)

# Get the indices for the split
for train_idx, test_idx in gss.split(X, y, groups):
    if isinstance(X, pd.DataFrame):
        X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
    else:
        X_train, X_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]

    neigh.fit(X_train, y_train)
    y_pred = neigh.predict(X_test)
    print(neigh.score(X_test, y_test))

print(X_test.shape)
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test, y_pred)
print(report)
```

```python
from sklearn.metrics import confusion_matrix
cnf_matrix = confusion_matrix(y_test, y_pred)
print('confusion matrix = ')
print(cnf_matrix)
```

```python
n_classes = 2
TP = np.zeros(n_classes)
FP = np.zeros(n_classes)
TN = np.zeros(n_classes)
FN = np.zeros(n_classes)
for c in range(n_classes) :
  TP[c] = sum(np.logical_and(y_test == c, y_pred == c))
  FP[c] = sum(np.logical_and(y_test != c, y_pred == c))
  TN[c] = sum((y_test != c) & (y_pred != c))
  FN[c] = sum([x and y for (x, y) in zip(y_test == c, y_pred !=c )])

print(f'{TP[0]=:.0f}, {TN[0]=:.0f}, {FP[0]=:.0f}, {FN[0]=:.0f}')
print(f'{TP[1]=:.0f}, {TN[1]=:.0f}, {FP[1]=:.0f}, {FN[1]=:.0f}')

pos_class = 1
TP, TN, FP, FN = TP[pos_class], TN[pos_class], FP[pos_class], FN[pos_class]
```

```python
#bitwise AND works well with Booleans actually
#I wouldn't really AND two numbers in ML
a = 8
b = 5
c = a & b
print(c)
```

```python
#logical_and returns b if a is not 0 or False.
#AND reads much better when coded as "and" rather than "&" in ML
a = 8
b = 5
c = a and b
print(c)
```

```python
#Pairs of scores like precision-recall, sensitivity-specificity, and TPR-TNR
#are better than using just the accuracy, especially with imbalanced data

P = sum(y_test == pos_class) #number of positive examples
sensitivity = recall = TPR = TP / P

N = n_test - P #number of negative examples
specificity = TNR = TN / N

precision = TP / (TP+FP)
```

```python
#Compute and print the metrics that appear in the classification report
from scipy.stats import hmean as harmonic_mean
F1_score = harmonic_mean([precision, recall])
print(f'{precision=}, {recall=}, {F1_score=}')

#All these metrics are also available in scikit-learn of course
from sklearn.metrics import precision_score, recall_score, f1_score
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
F1_score = f1_score(y_test, y_pred)
print(f'{precision=}, {recall=}, {F1_score=}')
```
