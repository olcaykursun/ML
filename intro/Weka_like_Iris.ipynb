```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


# Load iris data
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['class'] = iris.target

# Display the first few rows
data.head(20)
```

```python
# Plot histograms for each feature
data.drop(columns='class').hist(figsize=(12, 10), color='skyblue', edgecolor='black')
plt.tight_layout()  # Adjusts subplots for better layout
plt.show()
```

```python
#incorrect version for histogram of a feature showing the class distribution

# Choose 'sepal length (cm)' feature
feature = 'sepal length (cm)'

# Plotting histograms for each class on the same figure
colors = ['red', 'blue', 'green']
labels = iris.target_names

# Plot for each class
for class_num, color, label in zip([0, 1, 2], colors, labels):
    subset = data[data['class'] == class_num]
    plt.hist(subset[feature], bins=2, color=color, alpha=0.5, label=label)

plt.xlabel(feature)
plt.ylabel('Count')
plt.legend(loc='upper right')
plt.title(f'Histogram of {feature} by class')
plt.show()
```

```python
#Correct version for histogram of a feature showing the class distribution

# Choose 'sepal length (cm)' feature
feature = 'sepal length (cm)'

# Plotting histograms for each class on the same figure
colors = ['red', 'blue', 'green']
labels = iris.target_names

feat = data[feature]
counts, bins = np.histogram(feat, bins=2)

print(bins)

# Plot for each class
for class_num, color, label in zip([0, 1, 2], colors, labels):
    subset = data[data['class'] == class_num]
    plt.hist(subset[feature], bins=bins, color=color, alpha=0.5, label=label)

plt.xlabel(feature)
plt.ylabel('Count')
plt.legend(loc='upper right')
plt.title(f'Histogram of {feature} by class')
plt.show()
```

```python
import numpy as np

features = iris.feature_names
for i, feature_i in enumerate(features):
    for j, feature_j in enumerate(features):
        if i < j:  # To avoid duplicate plots and diagonal plots
            plt.figure(figsize=(6, 5))
            for class_num, marker, color in zip([0, 1, 2], ['o', 's', 'D'], ['red', 'blue', 'green']):
                subset = data[data['class'] == class_num]
                plt.scatter(subset[feature_i], subset[feature_j], color=color, marker=marker, label=iris.target_names[class_num])
            plt.xlabel(feature_i)
            plt.ylabel(feature_j)
            plt.legend()
            plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Split data into training and test sets
X = data.drop(columns='class')
y = data['class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)

# Use k-NN (e.g., with k=3)
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

# Predict and calculate accuracy
y_pred = knn.predict(X_test)
```

```python
accuracy = sum(y_test == y_pred)/len(y_test)
print(f"Accuracy: {accuracy * 100:.2f}%")

accuracy = (y_test == y_pred).mean()
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
from sklearn.utils.fixes import sklearn
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
